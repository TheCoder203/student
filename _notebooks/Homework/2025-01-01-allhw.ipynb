```python
%%js

class CookieClicker {
  constructor(cookies,cookiesPerClick, cookieType) {

    //Part 1.cookies and cookiesPerClick variables
    this.cookies = cookies; //start value
    this.cookiesPerClick = cookiesPerClick;
     this.cookieType = cookieType; //Step 5. Create cookie types
  }

  click() {
    //Part 2. Define what happens when cookie is clicked
    
    this.cookies += this.cookiesPerClick;
     console.log("You clicked a cookie!");
     console.log("You clicked a " + this.cookieType + " cookie!"); //Part 5. Type of cookie you clicked
    console.log("You have " + this.cookies + " cookies.");
  }

  //Part 3. Create a upgrade class that mutliplies cokkies per click
 applyUpgrade(upgrade) {
    this.cookiesPerClick *= upgrade.multiplier;
      // Part 4: Show change in cookie output
    console.log("Upgrade applied! Multiplier: " + upgrade.multiplier);
    this.showStatus(); 
  }

   showStatus() {
    console.log("Each click gives " + this.cookiesPerClick + " cookies. Total: " + this.cookies);
  }
}

class Upgrade {
  constructor(multiplier) {
    this.multiplier = multiplier;
  }
}


console.log("Starting CookieClicker Game...");
const game = new CookieClicker(0, 1, "Chocolate chip");
console.log("Initial setup complete.\n");


console.log("Clicking cookies...");
game.click();
game.click();
console.log("First round complete.\n");


console.log(" Applying upgrade...");
const upgrade = new Upgrade(2);
game.applyUpgrade(upgrade);
console.log("Upgrade applied.\n");


console.log("Clicking cookies after upgrade...");
game.click();
console.log("Final click complete.\n");


console.log(" Final Game Status:");
game.showStatus();
```

# Iterations

```python
%%js

function countStars(limit) {
  // 1. Initialize a counter variable. We'll start our count at 1.
  let count = 1;


  // 2. Start the `while` loop. The code inside this loop will run
  // as long as the condition `count <= limit` is true.
  while (count <= limit) {
    // 3. This is the code that runs in each iteration.
    console.log(`Star #${count} shining bright!`);


    // 4. This is the increment step. It's crucial! If we don't
    // increase the `count`, the loop will run forever (an infinite loop).
    count += 1
  }


  // 5. This message is displayed once the loop condition is no longer true
  // and the loop has finished.
  console.log("All the stars have been counted.");
}
```

```python
%%js

class Person {
  constructor(name, age) {
    this.name = name;
    this.age = age;
  }

  greet() {
    console.log(`Hello, my name is ${this.name}!`);
  }
}

// 3, 4 & 5) Book class with properties, read method, and isLongBook method
class Book {
  constructor(title, author, pages) {
    this.title = title;
    this.author = author;
    this.pages = pages;
    this.pagesRead = 0;
  }

  read(pages) {
    this.pagesRead += pages;
    if(this.pagesRead > this.pages) this.pagesRead = this.pages;
    console.log(`You have read ${this.pagesRead} out of ${this.pages} pages.`);
  }

  isLongBook() {
    return this.pages > 300;
  }
}

// 6 & 7) Car class with drive, refuel, and static info methods
class Car {
  constructor(make, model, fuel) {
    this.make = make;
    this.model = model;
    this.fuel = fuel; // assume fuel is in liters
  }

  drive(distance) {
    // Simple logic: driving consumes 0.1 liters per km
    const fuelNeeded = distance * 0.1;
    if (this.fuel >= fuelNeeded) {
      this.fuel -= fuelNeeded;
      console.log(`Drove ${distance} km. Fuel left: ${this.fuel.toFixed(2)} liters.`);
    } else {
      console.log("Not enough fuel to drive that distance.");
    }
  }

  refuel(amount) {
    this.fuel += amount;
    console.log(`Refueled ${amount} liters. Total fuel: ${this.fuel.toFixed(2)} liters.`);
  }

  static info() {
    console.log("Cars are vehicles.");
  }
}

// 8) ElectricCar extending Car with battery and charge method
class ElectricCar extends Car {
  constructor(make, model, fuel, battery) {
    super(make, model, fuel);
    this.battery = battery; // battery level in percentage 0-100
  }

  charge(amount) {
    this.battery += amount;
    if (this.battery > 100) this.battery = 100;
    console.log(`Battery charged to ${this.battery}%.`);
  }
}

// 10) Routine combining all actions
const person = new Person("Alice", 30);
const book = new Book("JavaScript Basics", "Anish Gupta", 350);
const car = new Car("Toyota", "Corolla", 10);
const electricCar = new ElectricCar("Tesla", "Model 3", 0, 50);

// Outputs
console.log("Person properties:", person);
person.greet();

console.log("\nBook properties:", book);
book.read(120);
console.log(`Is the book long? ${book.isLongBook()}`);

console.log("\nCar actions:");
car.drive(50);
car.refuel(5);

console.log("\nElectric Car actions:");
electricCar.charge(30);

console.log("\nStatic Car info:");
Car.info();
```
